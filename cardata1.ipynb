```python
df['Spare key']=df['Spare key'].map({"No":0,"Yes":1})
```

```python
import matplotlib.pyplot as plt
```

```python
x=df['Engine capacity']
y=df['Price']
plt.plot(x,y)
plt.title("Engine capacity vs Price ")
plt.xlabel("Engine capacity")
plt.ylabel("Price")
plt.show()
```

```python
df=df.sort_values(by='Manufacturing_year')
df_grouped=df.groupby("Manufacturing_year")["Price"].mean().reset_index()
```

```python
plt.plot(df_grouped['Manufacturing_year'],df_grouped["Price"],marker = 'o')
plt.show()
```

```python
df=df.sort_values(by='Engine capacity')
df_grouped1=df.groupby("Engine capacity")["Price"].mean().reset_index()
```

```python
plt.plot(df_grouped1['Engine capacity'],df_grouped1["Price"],marker = 'o')
plt.show()
```

```python
df_groupedk=df.groupby("Spare key")["Price"].mean().reset_index()
plt.bar(df_groupedk["Spare key"],df_groupedk["Price"],width = 0.1)
plt.title("Key vs Price")
plt.show()
```

```python
plt.hist(df['Price'],  color='skyblue', edgecolor='black')
plt.axvline(df['Price'].mean(), color='k', linestyle='dashed', linewidth=1,label="Mean") # to get the mean from histgram
plt.axvline(df['Price'].median(), color='blue', linewidth=1,label="Median") #to get the median from histgram 
plt.legend()
plt.show()
```

```python

plt.scatter(df['Imperfections'],df['Repainted Parts'] ,s=df['Price']/50000)
plt.title("Imperfections vs Repainted Parts") 
plt.xlabel("Imperfections")
plt.ylabel("Repainted Parts")
plt.show()
```

```python
count=df['Transmission'].value_counts().reset_index()
```

```python
plt.pie(count["count"],labels=count['Transmission'],autopct="%1.1f%%",startangle=90,explode=[0.1,0])
plt.title("Manual vd Automatic")
plt.show()
```

```python
countf=df['Fuel type'].value_counts().reset_index()
```

```python
#dount chart
plt.pie(countf["count"],labels=countf['Fuel type'],autopct="%1.1f%%",startangle=90,wedgeprops={"edgecolor":"black","width":0.5})
plt.title("Fuel type")
plt.show()
```

```python
plt.boxplot(
    df[['Engine capacity','Price']].values, 
    whis=[0,100],
    tick_labels =df[['Engine capacity','Price']].columns ,          
    patch_artist=True,         
    showmeans=True,            
    boxprops=dict(facecolor="lightblue"), 
    medianprops=dict(color="red"),         
    whiskerprops=dict(color="black"),      
    capprops=dict(color="black"),
    
)
plt.yscale('log') # Apply log scale to fix skew

plt.title("Box Plot of Engine capacity vs Price ")  
plt.ylabel("Engine capacity")     
plt.grid(True, linestyle="--", alpha=0.4)

plt.show()
```

```python
import seaborn as sns
corr = df.corr(numeric_only=True)
sns.heatmap(
    corr,
    annot=True,
    cmap="rocket_r",
    vmin=-1 ,vmax=1,
    linewidths=0.2
)
plt.title("corr heatmap")
plt.show()
```
