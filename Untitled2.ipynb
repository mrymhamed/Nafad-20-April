```python
import numpy as np
pid= np.array([101,102,103,104,105])
heart_rate = np.array([72,110,95,139,88])
#arr=np.array([pid,heart_rate]).T
#print(arr)
case = np.where(heart_rate>100,"Critical","Stable")
print(pid)
print(heart_rate)
print(case)
```

```python
temp = np.array([36.5,37.2,0,38.1,0,36.8])
condition = temp!=0
filteed=temp[condition]
temp_ave=filteed.mean()
cleaned_temp = np.where(temp ==0,temp_ave,temp)
print("Data before cleened: ",temp)
print("Data After cleened: ",cleaned_temp)
large37 = cleaned_temp>37
print(cleaned_temp[large37])
```

```python
age = np.array([25,65,45,70,50])
bp = np.array([120,150,135,160,145])
risk = np.where((age>50)|(bp>140),"High Risk","Stable")
print(risk)
```

```python
salary = np.array([400,600,550,300,700])
bonus = np.where(salary>=500,salary+100,salary+50)
print(bonus)
```

```python
mark = np.array([90,75,85,60,68])
grade = np.where(mark>85,"A",np.where(mark==85,"-A",np.where(mark>=70,"B","c")))
print(grade)
```

```python
purchase = np.array([50,120,80,200,150])
offer = np.where(purchase>100,purchase*0.8,purchase*0.9)
print("The original price is :",purchase)

print("After dicsount",offer)
```
