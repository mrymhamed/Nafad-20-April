```python
#Q7. Risk Detection (Simple Rule)
cholesterol = np.array([200,240,180,220,260])
pb=np.array([120,140,110,130,150])
high_risk = np.where((cholesterol>140)|(pb>240),"High risk","stable")
print(high_risk)
```

```python
#Compare Groups
under50 = patients_data[patients_data[:, 0] < 50]
print(under50)
print("\n")
above50 = patients_data[patients_data[:, 0] >= 50]
print(above50)
avg_under50 = np.mean(under50[:, 1])
avg_above50 = np.mean(above50[:, 1])

print("Avg heart rate (under 50):", avg_under50)
print("Avg heart rate (50 and above):", avg_above50)
```

```python
#Extreme Cases
Highest_blood_pressure = np.max(patients_data[:,3])
print(Highest_blood_pressure)
Lowest_heart_rate =np.min(patients_data[:,1])
print(Lowest_heart_rate)
```

```python
# Relationship Thinking
age = patients_data[:, 0]
bp = patients_data[:, 1]
younger_bp = bp[age < 50]
older_bp = bp[age >= 50]

print("Younger avg BP:", np.mean(younger_bp))
print("Older avg BP:", np.mean(older_bp))
print("Older patients show higher average blood pressure than younger ones")
```
